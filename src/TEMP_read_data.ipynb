```python
import geopandas as gpd
import pandas as pd
import os
```

```python
# Load the shapefile using all components
file_path = "../data/MOCT_link_model_input"
shp_path  = file_path + ".shp"
df_raw    = gpd.read_file(shp_path, encoding="cp949")

# Display the first few rows of the GeoDataFrame
df_raw.head()

#df_raw.to_csv(file_path + '.csv')
```

```python
cond_1 = (df_raw['F_NODE'] == '2250094600')
cond_2 = (df_raw['T_NODE'] == '2250094600')

df_raw[cond_1 | cond_2]

df_raw[(df_raw['LINK_ID'] == '2210002300') | (df_raw['LINK_ID'] == '2210002200')]
```

```python
# Load the uploaded CSV file
df_raw = pd.DataFrame(df_raw)

# Build edge connections: LINK_ID_1 and LINK_ID_2 where T_NODE of one matches F_NODE of another
df_mrg = pd.merge(
    df_raw[['LINK_ID', 'F_NODE', 'T_NODE']],
    df_raw[['LINK_ID', 'F_NODE', 'T_NODE']],
    left_on='T_NODE', right_on='F_NODE',
    suffixes=('_1', '_2')
)

# Keep only LINK_ID_1 and LINK_ID_2
df_link = df_mrg[['LINK_ID_1', 'LINK_ID_2']]

# Drop duplicates if needed
df_link = df_link.drop_duplicates()


df_link.head()

```
